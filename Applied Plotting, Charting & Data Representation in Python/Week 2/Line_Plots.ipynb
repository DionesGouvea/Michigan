```python
import matplotlib.pyplot as plt
import numpy as np

linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2

plt.figure()

#plotar o linear_data e o exponential_data
plt.plot(linear_data, '-o', exponential_data, '-o')

```

```python
#plotar outra serie com linha Dashed e cor vermelha 
plt.plot([22,44,55], '--r')
```

```python
#vamos adicionar as labels, titulo e legendas 
plt.plot(linear_data, '-o', exponential_data, '-o')
plt.plot([22,44,55], '--r')

plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
plt.legend(['Baseline', 'Competition', 'Us'])

```

```python
#preenche a Ã¡rea entre os dados lineares e os dados exponenciais
plt.plot(linear_data, '-o', exponential_data, '-o')
plt.gca().fill_between(range(len(linear_data)), 
                       linear_data, exponential_data, 
                       facecolor='blue', 
                       alpha=0.25)
```

Vamos tentar trabalhar com Datas 

```python
plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')
```

```python
#vamos tentar usar o pandas para isso 
import pandas as pd

plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) # resultara em erro se nÃ£o estiver dentro de uma lista 
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

```

```python
#as datas estÃ£o se sobrepondo vamos arrumar isto 

plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) # resultara em erro se nÃ£o estiver dentro de uma lista 
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

x = plt.gca().xaxis

# Vamos rotacionar as label do eixo X para elas nÃ£o se sobreporem 
for item in x.get_ticklabels():
    item.set_rotation(45)
 #   plt.subplots_adjust(bottom=0.25) aumentar o espaÃ§o na parte inferior nÃ£o necesseario aqui

#dando nome para labels, e titulo
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')
ax.set_title("Exponential ($x^2$) vs. Linear ($x$) performance")

```
