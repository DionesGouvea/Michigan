```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#vendo quais estilos podemos usar para mudar a aparencia dos nossos graficos
plt.style.available
```

```python
#usando o estilo seaborn
plt.style.use('seaborn-colorblind')
```

```python
#criando um dataframe com 3 colunas de valores aleatorios para testarmos visualização 
np.random.seed(123)

df = pd.DataFrame({'A': np.random.randn(365).cumsum(0), 
                   'B': np.random.randn(365).cumsum(0) + 20,
                   'C': np.random.randn(365).cumsum(0) - 20}, 
                  index=pd.date_range('1/1/2017', periods=365))
df.head()
```

```python
df.plot(); 
```

```python
#Podemos selecionar qual gráfico queremos usar passando-o para o parâmetro 'kind'.
df.plot('A','B', kind = 'scatter');
```

Voce também pode plotar escolhendo o tipo de grafico usando dataframe.plot.kind onde kind seria o tipo de grafico 

  kind :

- 'line' : line plot (default)
- 'bar' : vertical bar plot
- 'barh' : horizontal bar plot
- 'hist' : histogram
- 'box' : boxplot
- 'kde' : Kernel Density Estimation plot
- 'density' : same as 'kde'
- 'area' : area plot
- 'pie' : pie plot
- 'scatter' : scatter plot
- 'hexbin' : hexbin plot

```python
# criando um scatterplot da coluna a e c e mudando a cor e tamanho com base na coluna B
#C = color, S = size 
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
```

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
ax.set_aspect('equal')# mudando o aspecto do grafico
```

```python
#plotando boxplot
df.plot.box();
```

```python
#plotando um histograma
df.plot.hist(alpha=0.7);
```

```python
#grafico de densidade
df.plot.kde();
```
