```python
import pandas as pd
import matplotlib.pyplot as plt

pobres = pd.read_excel('/content/ipeadata[08-11-2022-11-13]- pobres.xls')
media = pd.read_excel('/content/ipeadata[08-11-2022-11-12] - media.xls')
ricos = pd.read_excel('/content/ipeadata[08-11-2022-10-50] - ricos .xls')

ricos = ricos.T
ricos.reset_index(inplace = True)
ricos.drop([0,1,2],inplace = True)

media = media.T
media.reset_index(inplace = True)
media.drop([0,1,2],inplace = True)

pobres = pobres.T
pobres.reset_index(inplace = True)
pobres.drop([0,1,2],inplace = True)
```
