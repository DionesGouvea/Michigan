```python
import pandas as pd
import matplotlib.pyplot as plt

#importando datasets 
media = pd.read_excel('/content/ipeadata[08-11-2022-11-12] - media.xls')
educa = pd.read_excel('/content/ipeadata[09-11-2022-12-26].xls')

#invertendo os eixos e me livrando de linhas que nao irei usar 
media = media.T
media.reset_index(inplace = True)
media.drop([0,1,2],inplace = True)

educa = educa.T
educa.reset_index(inplace = True)
educa.drop([0,1,2],inplace = True)
```

```python
#convertendo a coluna index para inteiro 
media['index'] = pd.to_numeric(media['index'])
media = media[media['index'] >= 1981] #pegando apenas os dados de 1981 para frente
```

```python
fig, ax1 = plt.subplots(figsize=(20,8))

#plot da taxa de analfabetismo
ax1.plot(educa['index'],educa[0],'-o',linewidth =3,label = 'percentage of illiterate people')
plt.yticks(color = 'b')
plt.ylabel('percentage of illiterate people', color = 'b',fontsize=18)

#plot da renda media da população brasileira. 
ax2 = ax1.twinx()
ax2.plot(educa['index'],media[0],'-o',color='r',linewidth =3,label ="Brazilian middle income")
plt.yticks(color = 'r')
plt.ylabel(" Brazilian middle income (in R$)", color = 'r',fontsize=18)


plt.title("Brazilian middle Income x Illiteracy rate 1981 - 2014", fontsize = 18)
plt.grid()


```
