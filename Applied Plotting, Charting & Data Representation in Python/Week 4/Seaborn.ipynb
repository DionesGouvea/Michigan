```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000), name='v2')
```

```python
plt.figure()
plt.hist(v1, alpha=0.7, bins=np.arange(-50,150,5), label='v1');
plt.hist(v2, alpha=0.7, bins=np.arange(-50,150,5), label='v2');
plt.legend();
```

```python
# plotar uma linha de densidade por cima do histograma
plt.figure()
plt.hist([v1, v2], histtype='barstacked',density=True);
v3 = np.concatenate((v1,v2))
sns.kdeplot(v3);
```

```python
plt.figure()
# podemos passar para cada componente chaves com argumentos individuais 
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'});
```

```python
# o joinplot serve para plotar as distribuições das variaveis ao mesmo tempo 
sns.jointplot(v1, v2, alpha=0.4);
```

```python
grid = sns.jointplot(v1, v2, alpha=0.4);
grid.ax_joint.set_aspect('equal') #mudando o aspecto de visualização 
```

```python
sns.jointplot(v1, v2, kind='hex'); #plotando um hexplot com as distribuições dos eixos 
```

```python
# definindo o estilo seaborn para os seguintes plots 
sns.set_style('white')

sns.jointplot(v1, v2, kind='kde', space=0);
```
