```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# criar uma grade de 2 linhas e 2 colunas 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)#compartilhando o eixo x
axs = [ax1,ax2,ax3,ax4]

#plotar 10, 100, 1000, e 10000 amostras de uma distribuição normal 
for n in range(0,len(axs)):
    sample_size = 10**(n+1) #tamanho da amostra 
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size) #gerando os valores aleatoriamente
    axs[n].hist(sample) #plotando a distribuição 
    axs[n].set_title('n={}'.format(sample_size)) #titulo
```

podemos perceber que conforme o numero de amostras e maior nossas barras do histograma ficam mais largas pois o eixo x é o mesmo para os 4, então a uma distribuição maior de dados para uma barra quando se possui mais amostras


```python
# Repetindo o exemplo porem agora setando o numero de bins para 100
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title('n={}'.format(sample_size))
```

com esse explo fica mais claro perceber quando olhamos para n = 10 ou 100

```python
#usando gridspec para particionar uma figura em subplots
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3) #criando uma grade de 3 colunas e 3 linhas 

#criando subplots e dizendo qual localização deles na grade 
top_histogram = plt.subplot(gspec[0, 1:]) 
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])

#gerando as distribuição
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)

lower_right.scatter(X, Y) # fazendo um scatterplot 
top_histogram.hist(X, bins=100) #um histograma
s = side_histogram.hist(Y, bins=100, orientation='horizontal') #um histograma com orientação horizontal
```
