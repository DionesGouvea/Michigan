```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
# subplot com 2 linhas 1 coluna 
plt.subplot(2, 1, 1) # o ultimo parametro e referente a numeração da figura 

#criando uma linha linear 
linear_data = np.array([1,2,3,4,5,6,7,8])
plt.plot(linear_data, '-o')
plt.text(0.5, 0.5, 'subplot(2,1,1)', ha='center', va='center',size=24, alpha=.5)

 

plt.subplot(2, 1, 2) #figura 2     
#criando uma linha exponencial 
exponential_data = linear_data**2 
plt.plot(exponential_data, '-o')
plt.text(0.5, 0.5, 'subplot(2,1,2)', ha='center', va='center',size=24, alpha=.5)
         

# plotando a linha exponencial na figura 1 
plt.subplot(2, 1, 1)
plt.plot(exponential_data, '-x')
```

```python
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, '-o')
#passando o sharey = ax1 para que as duas figuras tenham o mesmo tamanho no eixo Y
ax2 = plt.subplot(1, 2, 2, sharey=ax1)
plt.plot(exponential_data, '-x')
```

note que a linha linear esta quase deitada pois esta usando o eixo Y da linha exponencial
