###Descrição das Features 

**title_word_count** - o número de palavras no título do vídeo.

**document_entropy** - uma pontuação que indica a variedade de tópicos abordados no vídeo, com base na transcrição. Vídeos com pontuações de entropia menores tendem a ser mais coesos e mais focados em um único tópico.

**freshness** - Número de dias decorridos entre 01/01/1970 e a data de publicação da palestra. Os vídeos mais recentes terão valores de atualização mais altos.

**easiness** - Uma medida de dificuldade de texto aplicada à transcrição. Uma pontuação mais baixa indica linguagem mais complexa usada pelo apresentador.

**fraction_stopword_presence**  - Uma palavra irrelevante é uma palavra muito comum como 'o' ou 'e'. Esse recurso calcula a fração de todas as palavras que são stopwords na transcrição da palestra em vídeo.

**speaker_speed** - A taxa média de fala em palavras por minuto do apresentador no vídeo.

**silent_period_rate** - A fração de tempo no vídeo da palestra que é silêncio (sem fala).
