```python
#vamos começar trazendo nossas bibliotecas de processamento
import pandas as pd
import numpy as np

#também vamos pegar uma biblioteca de tempo 
import timeit

#e vamos pegar um dataframe de Ceus dos US
df = pd.read_csv('/content/census.csv')
df.head()
```

```python
# A primeira das expressões idiomáticas de pandas sobre as quais eu gostaria de falar é chamada de encadeamento de métodos. A ideia geral por trás
# encadeamento de métodos é que cada método em um objeto retorna uma referência a esse objeto. A beleza disso é
# que você pode condensar muitas operações diferentes em um DataFrame, por exemplo, em uma linha ou pelo menos uma
# declaração de código.

# Aqui está a maneira pandorável de escrever código com encadeamento de métodos. Neste código eu vou retirar o estado
# e nomes de cidades como um índice múltiplo, e farei isso apenas para dados que tenham um nível de resumo de 50,
# que neste conjunto de dados são dados em nível de condado. Vou renomear uma coluna também, apenas para torná-la um pouco mais legível

(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME','CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))
```

```python
# Vamos entender o que foi feito. Primeiro, usamos a função where() no dataframe e passamos uma máscara booleana que
# só é verdadeiro para as linhas em que SUMLEV é igual a 50. Isso indica em nossos dados de origem que os dados
# é resumido em nível de condado. Com o resultado da função where() avaliada, eliminamos
# valores. Lembre-se de que .where() não descarta valores ausentes por padrão. Em seguida, definimos um índice no resultado de
# este. Neste caso, configurei para o nome do estado seguido pelo nome do condado. Finalmente. Eu renomeio uma coluna para
# torná-lo mais legível. Note que ao invés de escrever tudo isso em uma linha, como eu poderia ter feito, eu comecei o
# declaração com um parêntese, que diz ao python que vou estender a declaração em várias linhas para
# legibilidade.

```

```python
# Aqui está uma maneira mais tradicional e não-pandorável de escrever isso. Não há nada de errado com este código no
# sentido funcional, você pode até entender melhor como uma nova pessoa no idioma. É apenas
# não é tão pandorável quanto o primeiro exemplo.

df = df[df['SUMLEV']==50] # Vou usar o operador de indexação sobrecarregado [] que descarta nans 
# inserindo um novo index para nosso dataframe e passando inplace para dizer que é uma mudança permanente 
df.set_index(['STNAME','CTYNAME'], inplace=True)
# renomeando nome de coluna 
df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

```python
# Agora, a chave com qualquer boa expressão idiomática é entender quando ela não está ajudando você. Neste caso, você pode realmente
# cronometra ambos os métodos e vê qual deles roda mais rápido

# Podemos colocar a abordagem em uma função e passar a função para a função timeit para contar o tempo que o
# número do parâmetro nos permite escolher quantas vezes queremos executar a função. Aqui vamos apenas configurá-lo para
# 10
def first_approach():
    global df #nossa variavel pode ser usada fora da função 
    # agora colamos nosso codigo 
    return (df.where(df['SUMLEV']==50)
             .dropna()
             .set_index(['STNAME','CTYNAME'])
             .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))

#vamos ler o nosso conjunto de dados novamente 
df = pd.read_csv('/content/census.csv')

#agora vamos rodar isso 
timeit.timeit(first_approach, number=10)
```

```python
# Agora vamos testar a segunda abordagem. Como você pode notar, usamos nossa variável global df na função.
# No entanto, alterar uma variável global dentro de uma função modificará a variável mesmo em um escopo global e nós
# não quero que isso aconteça neste caso. Portanto, para selecionar apenas níveis de resumo de 50, crio um novo
# dataframe para esses registros

def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME','CTYNAME'], inplace=True)
    return new_df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})

df = pd.read_csv('/content/census.csv')

timeit.timeit(second_approach, number=10)
```

```python
# Como você pode ver, a segunda abordagem é muito mais rápida! Então, este é um exemplo particular de um tempo clássico
# troca de legibilidade.

# Você verá muitos exemplos sobre estouro de pilha e na documentação de pessoas usando encadeamento de métodos em seus
#panda. E assim, acho que ser capaz de ler e entender a sintaxe realmente vale o seu tempo. Mas mantenha-se
# lembre-se de que seguir o que parece ser expressões idiomáticas estilísticas pode ter problemas de desempenho que você precisa
# considere também.

```

```python
# Vamos dar uma olhada em nosso censo DataFrame. Neste DataFrame, temos cinco colunas para estimativas populacionais,
# com cada coluna correspondendo a um ano de estimativas. É bastante razoável querer criar alguns novos
# colunas para valores mínimos ou máximos, e a função apply é uma maneira fácil de fazer isso.

# Primeiro, precisamos escrever uma função que receba uma determinada linha de dados, encontre um mínimo e um máximo
# valores e retorna uma nova linha de dados nd retorna uma nova linha de dados. Chamaremos esta função min_max

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})


#agora so precisamos aplicar o aply 
# vale lembrar quee se quisermos passar em todas as linhas de todas as colunas 
#temos que passar o eixo = colunas
df.apply(min_max, axis='columns').head()
```

```python
# Claro que não há necessidade de se limitar a retornar um novo objeto de série. Se você está fazendo isso como parte
# de limpeza de dados é provável que você queira adicionar novos dados ao DataFrame existente. Naquilo
# caso você apenas pegue os valores das linhas e adicione novas colunas indicando as pontuações máxima e mínima. Isto é um
# parte regular do meu fluxo de trabalho ao trazer dados e construir resumos ou estatísticas descritivas, e é
# frequentemente usado fortemente com a fusão de DataFrames.


#aqui vai um exemplo da função onde ela retorna duas novas colunas para inserir no dataframe
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    # criando uma nova coluna para max e min
    row['max'] = np.max(data)
    
    row['min'] = np.min(data)
    return row
# Agora vamos aplicar ao dataframe a nossa função 
df.apply(min_max, axis='columns')
```

```python
#aqui vai um exemplo onde vc pode encadear varias chamadas de apply para criar um  script de manipulação 
#mais sucinto. Um exemplo de uma linha onde você pode calcular o máximo das colunas
# usando a função aplicar.
rows = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013','POPESTIMATE2014', 
        'POPESTIMATE2015']

#agora vamos aplicar um apply com lambda (lembre que lambda e uma função sem nome de uma linha)
#neste caso ela recebe um único parâmetro, x, e retorna um único valor, neste
# case o máximo em todas as colunas associadas à linha x.
df.apply(lambda x: np.max(x[rows]), axis=1).head()
```

```python
# A beleza da função apply é que ela permite flexibilidade em fazer qualquer manipulação que você
# desejar, pois a função que você passa para apply pode ser qualquer customizada como você quiser. Digamos que queremos
# divide os estados em quatro categorias: Nordeste, Centro-Oeste, Sul e Oeste Podemos escrever um
# função que retorna a região com base no estado as informações das regiões do estado são obtidas da Wikipedia

def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"

```

```python
# Agora temos a função customizada, digamos que queremos criar uma nova coluna chamada Region, que mostra a
# região do estado, podemos usar a função customizada e a função apply para isso. A função personalizada
# deve funcionar na coluna de nome de estado STNAME. Então vamos definir a função apply no nome do estado
# coluna e passe a função customizada para a função apply

df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
#vamos ver o resultado 
df[['STNAME','state_region']].head()
```
