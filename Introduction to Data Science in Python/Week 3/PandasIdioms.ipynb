```python
#vamos comeÃ§ar trazendo nossas bibliotecas de processamento
import pandas as pd
import numpy as np

#tambÃ©m vamos pegar uma biblioteca de tempo 
import timeit

#e vamos pegar um dataframe de Ceus dos US
df = pd.read_csv('/content/census.csv')
df.head()
```

```python
# A primeira das expressÃµes idiomÃ¡ticas de pandas sobre as quais eu gostaria de falar Ã© chamada de encadeamento de mÃ©todos. A ideia geral por trÃ¡s
# encadeamento de mÃ©todos Ã© que cada mÃ©todo em um objeto retorna uma referÃªncia a esse objeto. A beleza disso Ã©
# que vocÃª pode condensar muitas operaÃ§Ãµes diferentes em um DataFrame, por exemplo, em uma linha ou pelo menos uma
# declaraÃ§Ã£o de cÃ³digo.

# Aqui estÃ¡ a maneira pandorÃ¡vel de escrever cÃ³digo com encadeamento de mÃ©todos. Neste cÃ³digo eu vou retirar o estado
# e nomes de cidades como um Ã­ndice mÃºltiplo, e farei isso apenas para dados que tenham um nÃ­vel de resumo de 50,
# que neste conjunto de dados sÃ£o dados em nÃ­vel de condado. Vou renomear uma coluna tambÃ©m, apenas para tornÃ¡-la um pouco mais legÃ­vel

(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME','CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))
```

```python
# Vamos entender o que foi feito. Primeiro, usamos a funÃ§Ã£o where() no dataframe e passamos uma mÃ¡scara booleana que
# sÃ³ Ã© verdadeiro para as linhas em que SUMLEV Ã© igual a 50. Isso indica em nossos dados de origem que os dados
# Ã© resumido em nÃ­vel de condado. Com o resultado da funÃ§Ã£o where() avaliada, eliminamos
# valores. Lembre-se de que .where() nÃ£o descarta valores ausentes por padrÃ£o. Em seguida, definimos um Ã­ndice no resultado de
# este. Neste caso, configurei para o nome do estado seguido pelo nome do condado. Finalmente. Eu renomeio uma coluna para
# tornÃ¡-lo mais legÃ­vel. Note que ao invÃ©s de escrever tudo isso em uma linha, como eu poderia ter feito, eu comecei o
# declaraÃ§Ã£o com um parÃªntese, que diz ao python que vou estender a declaraÃ§Ã£o em vÃ¡rias linhas para
# legibilidade.

```

```python
# Aqui estÃ¡ uma maneira mais tradicional e nÃ£o-pandorÃ¡vel de escrever isso. NÃ£o hÃ¡ nada de errado com este cÃ³digo no
# sentido funcional, vocÃª pode atÃ© entender melhor como uma nova pessoa no idioma. Ã‰ apenas
# nÃ£o Ã© tÃ£o pandorÃ¡vel quanto o primeiro exemplo.

df = df[df['SUMLEV']==50] # Vou usar o operador de indexaÃ§Ã£o sobrecarregado [] que descarta nans 
# inserindo um novo index para nosso dataframe e passando inplace para dizer que Ã© uma mudanÃ§a permanente 
df.set_index(['STNAME','CTYNAME'], inplace=True)
# renomeando nome de coluna 
df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

```python
# Agora, a chave com qualquer boa expressÃ£o idiomÃ¡tica Ã© entender quando ela nÃ£o estÃ¡ ajudando vocÃª. Neste caso, vocÃª pode realmente
# cronometra ambos os mÃ©todos e vÃª qual deles roda mais rÃ¡pido

# Podemos colocar a abordagem em uma funÃ§Ã£o e passar a funÃ§Ã£o para a funÃ§Ã£o timeit para contar o tempo que o
# nÃºmero do parÃ¢metro nos permite escolher quantas vezes queremos executar a funÃ§Ã£o. Aqui vamos apenas configurÃ¡-lo para
# 10
def first_approach():
    global df #nossa variavel pode ser usada fora da funÃ§Ã£o 
    # agora colamos nosso codigo 
    return (df.where(df['SUMLEV']==50)
             .dropna()
             .set_index(['STNAME','CTYNAME'])
             .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))

#vamos ler o nosso conjunto de dados novamente 
df = pd.read_csv('/content/census.csv')

#agora vamos rodar isso 
timeit.timeit(first_approach, number=10)
```

```python
# Agora vamos testar a segunda abordagem. Como vocÃª pode notar, usamos nossa variÃ¡vel global df na funÃ§Ã£o.
# No entanto, alterar uma variÃ¡vel global dentro de uma funÃ§Ã£o modificarÃ¡ a variÃ¡vel mesmo em um escopo global e nÃ³s
# nÃ£o quero que isso aconteÃ§a neste caso. Portanto, para selecionar apenas nÃ­veis de resumo de 50, crio um novo
# dataframe para esses registros

def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME','CTYNAME'], inplace=True)
    return new_df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})

df = pd.read_csv('/content/census.csv')

timeit.timeit(second_approach, number=10)
```

```python
# Como vocÃª pode ver, a segunda abordagem Ã© muito mais rÃ¡pida! EntÃ£o, este Ã© um exemplo particular de um tempo clÃ¡ssico
# troca de legibilidade.

# VocÃª verÃ¡ muitos exemplos sobre estouro de pilha e na documentaÃ§Ã£o de pessoas usando encadeamento de mÃ©todos em seus
#panda. E assim, acho que ser capaz de ler e entender a sintaxe realmente vale o seu tempo. Mas mantenha-se
# lembre-se de que seguir o que parece ser expressÃµes idiomÃ¡ticas estilÃ­sticas pode ter problemas de desempenho que vocÃª precisa
# considere tambÃ©m.

```

```python
# Vamos dar uma olhada em nosso censo DataFrame. Neste DataFrame, temos cinco colunas para estimativas populacionais,
# com cada coluna correspondendo a um ano de estimativas. Ã‰ bastante razoÃ¡vel querer criar alguns novos
# colunas para valores mÃ­nimos ou mÃ¡ximos, e a funÃ§Ã£o apply Ã© uma maneira fÃ¡cil de fazer isso.

# Primeiro, precisamos escrever uma funÃ§Ã£o que receba uma determinada linha de dados, encontre um mÃ­nimo e um mÃ¡ximo
# valores e retorna uma nova linha de dados nd retorna uma nova linha de dados. Chamaremos esta funÃ§Ã£o min_max

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})


#agora so precisamos aplicar o aply 
# vale lembrar quee se quisermos passar em todas as linhas de todas as colunas 
#temos que passar o eixo = colunas
df.apply(min_max, axis='columns').head()
```

```python
# Claro que nÃ£o hÃ¡ necessidade de se limitar a retornar um novo objeto de sÃ©rie. Se vocÃª estÃ¡ fazendo isso como parte
# de limpeza de dados Ã© provÃ¡vel que vocÃª queira adicionar novos dados ao DataFrame existente. Naquilo
# caso vocÃª apenas pegue os valores das linhas e adicione novas colunas indicando as pontuaÃ§Ãµes mÃ¡xima e mÃ­nima. Isto Ã© um
# parte regular do meu fluxo de trabalho ao trazer dados e construir resumos ou estatÃ­sticas descritivas, e Ã©
# frequentemente usado fortemente com a fusÃ£o de DataFrames.


#aqui vai um exemplo da funÃ§Ã£o onde ela retorna duas novas colunas para inserir no dataframe
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    # criando uma nova coluna para max e min
    row['max'] = np.max(data)
    
    row['min'] = np.min(data)
    return row
# Agora vamos aplicar ao dataframe a nossa funÃ§Ã£o 
df.apply(min_max, axis='columns')
```

```python
#aqui vai um exemplo onde vc pode encadear varias chamadas de apply para criar um  script de manipulaÃ§Ã£o 
#mais sucinto. Um exemplo de uma linha onde vocÃª pode calcular o mÃ¡ximo das colunas
# usando a funÃ§Ã£o aplicar.
rows = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013','POPESTIMATE2014', 
        'POPESTIMATE2015']

#agora vamos aplicar um apply com lambda (lembre que lambda e uma funÃ§Ã£o sem nome de uma linha)
#neste caso ela recebe um Ãºnico parÃ¢metro, x, e retorna um Ãºnico valor, neste
# case o mÃ¡ximo em todas as colunas associadas Ã  linha x.
df.apply(lambda x: np.max(x[rows]), axis=1).head()
```

```python
# A beleza da funÃ§Ã£o apply Ã© que ela permite flexibilidade em fazer qualquer manipulaÃ§Ã£o que vocÃª
# desejar, pois a funÃ§Ã£o que vocÃª passa para apply pode ser qualquer customizada como vocÃª quiser. Digamos que queremos
# divide os estados em quatro categorias: Nordeste, Centro-Oeste, Sul e Oeste Podemos escrever um
# funÃ§Ã£o que retorna a regiÃ£o com base no estado as informaÃ§Ãµes das regiÃµes do estado sÃ£o obtidas da Wikipedia

def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"

```

```python
# Agora temos a funÃ§Ã£o customizada, digamos que queremos criar uma nova coluna chamada Region, que mostra a
# regiÃ£o do estado, podemos usar a funÃ§Ã£o customizada e a funÃ§Ã£o apply para isso. A funÃ§Ã£o personalizada
# deve funcionar na coluna de nome de estado STNAME. EntÃ£o vamos definir a funÃ§Ã£o apply no nome do estado
# coluna e passe a funÃ§Ã£o customizada para a funÃ§Ã£o apply

df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
#vamos ver o resultado 
df[['STNAME','state_region']].head()
```
