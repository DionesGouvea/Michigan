```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
Energy = pd.read_excel('/content/Energy Indicators.xls',header = 17,nrows =227,
                   names =['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'],usecols =[2,3,4,5],
                   na_values='...',index_col = 0)
Energy.head(50)
```

```python
Energy['Energy Supply'] = Energy['Energy Supply'] * 1000000
#Renomeando o indice tirando os () e numeros do nome com regex e str.extract
Energy.index = Energy.index.str.extract('(?P<Country>[A-zÀ-ÿ\', ]+)', expand=False)

#renomeando o indice passando um dicionario 
Energy.rename(index = {"Republic of Korea": "South Korea",
                   "United States of America": "United States",
                   "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                   "China, Hong Kong Special Administrative Region": "Hong Kong",
                   "Iran ": 'Iran'},
          inplace = True)


Energy.head(100)
```

```python
#fazendo um for para pegar as colunas de 2006 a 2016
cols = [str(i) for i in range(2006,2016)]
#inserindo na posição 0 contry name para ser o index
cols.insert(0, 'Country Name')
GDP = pd.read_csv('/content/world_bank.csv', header = 4, index_col = 0, usecols = cols)

#renomeando alguns indices 
GDP.rename(index ={"Korea, Rep.": "South Korea", 
                  "Iran, Islamic Rep.": "Iran",
                  "Hong Kong SAR, China": "Hong Kong"}, inplace = True)
GDP.head(113)
```

```python
ScimEn = pd.read_excel('/content/scimagojr-3.xlsx',nrows =15,index_col=1)

def answer_one():
  #criando uma lista para ordenar as colunas 
  novaOrdem = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 
                          'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', 
                          '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', 
                          '2015']
  #fazendo um merge entre dois dataframes                         
  x = pd.merge(Energy, ScimEn, how = 'inner', on = 'Country')
  #atribuindo o Dataframe GDP para variavel y para poder fazer o merge sem alterar o dataframe original
  y = GDP
  y.index.name = 'Country'
  resposta = pd.merge(x,y,how ='inner', on = 'Country')

  #passando a nova ordem de colunas e ordenando os valores pela coluna Rank
  resposta = resposta[novaOrdem].sort_values('Rank')
  return resposta
  
answer_one()
```

# Questão 3

Esta função deve retornar uma série chamada avgGDP com 15 países e seu PIB médio classificado em ordem decrescente.

```python
def answer_three():
  avgGDP = answer_one()
  #usando um for para pegar os valores de cada ano, axis= 1 para dizer que e por coluna
  return avgGDP[[str(i) for i in range(2006, 2016)]].mean(axis = 1).sort_values(ascending = False)
answer_three()
```

# Questão 10
Crie uma nova coluna com 1 se o valor de % renovável do país for igual ou superior à mediana para todos os países no top 15 e 0 se o valor de % renovável do país estiver abaixo da mediana.

Essa função deve retornar uma série chamada HighRenew cujo índice é o nome do país classificado em ordem crescente de classificação.

```python
def answer_ten():
  mediana = answer_one()['% Renewable'].median()
  q = answer_one()
  i = []
  for x in answer_one()['% Renewable']:
    if x >= mediana:
      i.append(1)  
    else:
      i.append(0)
  q['HighRenew'] = i
  return q['HighRenew']

answer_ten()
```

# Questão 11 

Use o dicionário a seguir para agrupar os países por continente e, em seguida, crie um DataFrame que exiba o tamanho da amostra (o número de países em cada compartimento de continente) e a soma, média e desvio padrão para a população estimada de cada país.

    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

  
Esta função deve retornar um DataFrame com índice chamado Continente ['Ásia', 'Austrália', 'Europa', 'América do Norte', 'América do Sul'] e colunas ['size', 'sum', 'mean', 'std ']


```python
def answer_eleven():
  ContinentDict  = {'China':'Asia', 
              'United States':'North America', 
              'Japan':'Asia', 
              'United Kingdom':'Europe', 
              'Russian Federation':'Europe', 
              'Canada':'North America', 
              'Germany':'Europe', 
              'India':'Asia',
              'France':'Europe', 
              'South Korea':'Asia', 
              'Italy':'Europe', 
              'Spain':'Europe', 
              'Iran':'Asia',
              'Australia':'Australia', 
              'Brazil':'South America'}
  
  top = answer_one()
  top['Popula'] = top['Energy Supply'] / top['Energy Supply per Capita']
  z =[]
  for x in top.index:
    z.append(ContinentDict[x])

  top['Continent'] = z
  df = top
  h = pd.pivot_table(df, index = 'Continent', values = 'Popula', 
                          aggfunc=[np.size, np.sum, np.mean, np.std])
  h.columns = ['size', 'sum', 'mean', 'std']
  return h


answer_eleven()
```

```python
def answer_twelve():
  ContinentDict  = {'China':'Asia', 
              'United States':'North America', 
              'Japan':'Asia', 
              'United Kingdom':'Europe', 
              'Russian Federation':'Europe', 
              'Canada':'North America', 
              'Germany':'Europe', 
              'India':'Asia',
              'France':'Europe', 
              'South Korea':'Asia', 
              'Italy':'Europe', 
              'Spain':'Europe', 
              'Iran':'Asia',
              'Australia':'Australia', 
              'Brazil':'South America'}
  
  top = answer_one()
  top['Popula'] = top['Energy Supply'] / top['Energy Supply per Capita']
  z =[]
  for x in top.index:
    z.append(ContinentDict[x])
  top['Continent'] = z
  

  top['% Renewable']=pd.cut(top['% Renewable'],5)
  new_df=top.groupby(['Continent','% Renewable'])['Continent'].agg(np.size).dropna()
  return new_df

answer_twelve()
```
