Como vimos, tanto Series quanto DataFrames podem ter índices aplicados a eles. O índice é essencialmente um rótulo de nível de linha e, em pandas, as linhas correspondem ao eixo zero. Os índices podem ser gerados automaticamente, como quando criamos uma nova série sem um índice, caso em que obtemos valores numéricos, ou podem ser definidos explicitamente, como quando usamos o objeto de dicionário para criar a série, ou quando carregamos dados do arquivo CSV e defina os parâmetros apropriados. Outra opção para definir um índice é usar a função set_index(). Essa função pega uma lista de colunas e promove essas colunas a um índice. Nesta palestra, exploraremos mais sobre como os índices funcionam nos pandas.


```python
# A função set_index() é um processo destrutivo e não mantém o índice atual.
# Se você deseja manter o índice atual, você precisa criar manualmente uma nova coluna e copiar para
# ela os valores do atributo index.

#vamos chamar o nosso csv
import pandas as pd
df = pd.read_csv("/content/Admission_Predict.csv", index_col=0)
df.head()

```

```python
# Digamos que não queremos indexar o DataFrame por números de série, mas sim pelo
# chance de admitir. Mas vamos supor que queremos manter o número de série para mais tarde. Então vamos
# preserva o número de série em uma nova coluna. 

#copiando o index para esta coluna
df['Serial Number'] = df.index

# passando o chance of admit como novo index
df = df.set_index('Chance of Admit ')
df.head()
```

```python
#podemos criar um novo index com a função reset_index, ela vai criar um index com numeros padroes 
df = df.reset_index()
df.head()
```

```python
#vamos alterar nosso conjunto de dados para um exemplo diferente de indexação 
df = pd.read_csv('/content/census.csv')
df.head()
```

```python
#digamos que queremos dois indices um com o nome do estado e o segundo com o nome da cidade 
#nesse conjunto de dados temos dois tipos um tipo contendo informações de todo pais, e outro tipo contendo 
#informação de todo estado. 
#vamos confirmar que realmente so existe esses dois tipos de dados 
df['SUMLEV'].unique()
```

```python
#certo agora vamos excluir os dados de estado e manter os dados com a informação de pais 
df=df[df['SUMLEV']==50 ]
df.head()
```

```python
#vamos reduzir o numero de colunas e manter apenas as que nos interessa 

columns_to_keep = ['STNAME','CTYNAME','BIRTHS2010','BIRTHS2011','BIRTHS2012','BIRTHS2013',
                   'BIRTHS2014','BIRTHS2015','POPESTIMATE2010','POPESTIMATE2011',
                   'POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']
df = df[columns_to_keep]
df.head()
```

```python
#agora vamos criar nosso multi index o primeiro contendo o nome do estado e o segundo contendo as cidades
df = df.set_index(['STNAME', 'CTYNAME'])
df.head()
```

```python
#caso a gente queira pesquisar os dados de Washtenaw que fica no estado de michigan ficaria desta forma 
df.loc['Michigan', 'Washtenaw County']
```

```python
#caso a queiramos comparar 2 cidades tempos que passar uma tupla contendo o estado e cidade de ambos 
df.loc[ [('Michigan', 'Washtenaw County'),
         ('Michigan', 'Wayne County')] ]
```
