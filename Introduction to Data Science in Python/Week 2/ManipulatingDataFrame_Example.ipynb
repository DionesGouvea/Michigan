```python
import pandas as pd 
#vamos carregar nosso conjunto de dados 
df=pd.read_csv("/content/presidents.csv")
df.head()
```

```python
# Ok, temos alguns presidentes, algumas datas, vejo um monte de notas de rodapé na coluna "Nascido" que podem causar
# questões. Vamos começar limpando esse nome em nome e sobrenome.Eu vou lidar com isso com
# uma regex. Então eu quero criar duas novas colunas e aplicar uma regex na projeção da coluna "Presidente".

#vamos criar uma copia da coluna presidente e armazenar na coluna firt
df["First"]=df['President']

#agora vamos aplicar regex para remover o sobrenome 
df["First"]=df["First"].replace("[ ].*", "", regex=True)

df.head()
```

```python
# Isso funciona, mas é meio nojento. E é lento, pois tivemos que fazer uma cópia completa de uma coluna e depois ir
# através de strings de atualização. Existem algumas outras maneiras de lidar com isso. Deixe-me mostrar-lhe o mais

#primeiro vamos remover a coluna First
del(df["First"])

# A função apply() em um dataframe pegará alguma função arbitrária que você escreveu e a aplicará
# uma série (uma única coluna) ou DataFrame em todas as linhas ou colunas. Vamos escrever uma função que
# apenas divide uma string em duas partes usando uma única linha de dados

def splitname(row):
    #vamos pegar a primeira informação da coluna presidente e armazenar em um nova entrada
    row['First']=row['President'].split(" ")[0]
    # mesma coisa aqui mas iremos pegar o ultimo valor 
    row['Last']=row['President'].split(" ")[-1]
     
    return row

#agora vamos aplicar isto ao nosso dataframe indicando que queremos aplicar nas colunas
df=df.apply(splitname, axis='columns')
df.head()
```

```python
#vamos usar agora a função extract 
#primeiro vamos deletar as nossas colunas criadas 
del(df['First'])
del(df['Last'])

#extract pega uma expressão regular para podermos extrair aquilo que desejamos 
#vamos fazer uma expressão para pegar o primeiro e o ultimo nome 
pattern="(^[\w]*)(?:.* )([\w]*$)"

#agora vamos chamar nossa função extract usando Series.str.extract(pattern)
df["President"].str.extract(pattern).head()
```

```python
#podemos também extrair os dados e criar novas colunas para armazenar eles 

pattern="(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

#chamando a função 
names=df["President"].str.extract(pattern).head()
names
```

```python
#e podemos copiar isso para nosso main dataframe 
df["First"]=names["First"]
df["Last"]=names["Last"]
df.head()
```
