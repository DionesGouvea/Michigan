# DataFrame carregamento e indexação

```python
import pandas as pd
#podemos importar um conjunto de dados em formato csv para usar como dataFrame 

df = pd.read_csv('datasets/Admission_Predict.csv')

df.head()
```

```python
#podemos ver que o pandas criou um index que se inicia em zero, mas a nosso dataset
#ja possui um index bem definido, podemos pedir pro pandar não criar um index da seguinte forma
df = pd.read_csv('datasets/Admission_Predict.csv', index_col = 0)

df.head()
```

```python
#temos duas colunas com nomes SOP e LOR e para facilitar o entendimento dessas colunas 
#iremos renomealas 

new_df=df.rename(columns={'GRE Score':'GRE Score', 'TOEFL Score':'TOEFL Score',
                   'University Rating':'University Rating', 
                   'SOP': 'Statement of Purpose','LOR': 'Letter of Recommendation',
                   'CGPA':'CGPA', 'Research':'Research',
                   'Chance of Admit':'Chance of Admit'})
new_df.head()
```

```python
#note que apenas o SOP foi mudado de nome, vamos investigar o motivo 
#primeiro checamos se o nome das colunas estão corretos

new_df.columns
```

```python
#pdemos ver que existe um espaço apos o LOR e por isso não funcionou nosso rename
#vamos renomear usando o espaço desta ver 
new_df=new_df.rename(columns={'LOR ': 'Letter of Recommendation'})
new_df.head()
```

```python
#note que não colocamos o dicionario todo pois estamos apenas renomeando o mesmo dataset 
#anteriomente nos criamos outro e renomeamos ao mesmo tempo.
# se você notar chance de admitir também possui um espaço, as vezes pegamos datasets com dados 
# com duplo espaço, espaço no começo e etc, uma maneira mais eficas de padronizar os nomes antes de trabalhar
#com eles e usando uma função

new_df=new_df.rename(mapper=str.strip, axis='columns')
new_df.columns
```

```python
#veja que chance of admit agora nao possui mais espaço 
```

```python
#Bom ainda temos nosso Data Frame original e ele ainda possui os espaços no nomes das colunas 
df.columns
```

```python
#vamos supor que nos queiramos mudar todos os nomes das nossas colunas para letras minusculas 
#primeiro precisamos obter uma lista com o nome das nossas colunas 

cols = list(df.columns)

#agora usaremos uma função para fazer isso para nos 
cols = [x.lower().strip() for x in cols]

#agora vamos sobrescrever o nome das colunas no nosso DataFrame 
df.columns=cols
# vamos ver o resultado 
df.head()
```
