```python
import pandas as pd

#vamos carregar nosso arquivo csv
df = pd.read_csv('/content/Admission_Predict.csv',index_col = 0)
#vamos limpar os nomes das colunas como ja fizemos anteriormente 
df.columns = [x.lower().strip() for x in df.columns]
#vamos ver como ficou
df.head()
```

```python
#vamos criar um Boolean mask que ira armazenar os alunos que tem chance de ser admitidos

admit_mask=df['chance of admit'] > 0.7
admit_mask
```

```python
#mas o que podemos fazer com essa mask ? podemos chamar o nosso dataframe com um where e passar a nossa mask como filtro 
df.where(admit_mask).head()
```

```python
#vemos que os valores que não atendem o nosso filtro aparecem como NaN porem como fazemos para removelos do nosso dataframe?
#vamos usar a função dropna()
df.where(admit_mask).dropna().head()
```

```python
#como podemos ver o dado de numero 5 simiu do nosso dataset
#existe uma forma mais comum e simplificada de fazer isto vamos ver como funciona

df[df['chance of admit']> 0.7].head()
```

```python
#desta forma podemos chamar uma unica coluna 
df["gre score"].head()
```

```python
#ou chamar uma lista de colunas 
df[['gre score', 'toefl score']].head()
```

```python
#porem se quiseremos comparar dois testes boleanos como criterio para chamar nosso dataset 
#o pandas nao funciona muito bem por se tratar de uma series 
(df['chance of admit'] > 0.7) and (df['chance of admit'] < 0.9)
```

```python
#um jeito funcional de fazer isto é usando um & no lugar do and 
#é importante que usemos () para fazer essa comparação pois ao contrario ira gerar erro.
(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
```

```python
#outra forma de fazer isso e esconder nossos operadores condicionais e usar .gt e .lt
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```
