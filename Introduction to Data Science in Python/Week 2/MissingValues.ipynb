```python
import pandas as pd

#vamos abrir o nosso arquivo csv 
df = pd.read_csv('/content/class_grades.csv')
df.head(10)
```

```python
#podemos usar a função isnull para criar uma mask dos dados faltantes 
mask=df.isnull()
mask.head(10)
```

```python
# Isso pode ser útil para processar linhas com base em determinadas colunas de dados. Outra operação útil é ser
# capaz de eliminar todas as linhas que possuem dados ausentes, o que pode ser feito com a função dropna().
df.dropna().head(10)
```

```python
# Observe como as linhas indexadas com 2, 3, 7 e 11 desapareceram. Uma das funções úteis que o Pandas tem para
# trabalhando com valores ausentes é a função de preenchimento, fillna(). Esta função recebe um número ou parâmetros.
# Você pode passar um único valor que é chamado de valor escalar para alterar todos os dados ausentes para um
# valor. Isso não é realmente aplicável neste caso, mas é um caso de uso bastante comum.

# Então, se quiséssemos preencher todos os valores ausentes com 0, usaríamos fillna
df.fillna(0, inplace=True)
df.head(10)
```

```python
# Observe que o atributo inplace faz com que os pandas preencham os valores embutidos e não retorne uma cópia do
# dataframe, mas modifica o dataframe que você possui.
```

```python
#vamos importar um novo conjunto de dados que armazena o log de quem esta assistindo videos online 
#ele armazena o tempo, o nome de usuario, qual video, volume, posição e se esta pausado
df = pd.read_csv("/content/log.csv")
df.head(20)
```

```python
#vamos definir o nosso index como sendo time, e ordenar por ele 
df = df.set_index('time')
df = df.sort_index()
df.head(20)
```

```python
#podemos ver que temos mais de um usuario com o mesmo time então vamos fazer um multiindex para ficar melhor
df = df.reset_index()
df = df.set_index(['time', 'user'])
df
```

```python
#agora vamos preencher os dados fantantes com ffill
#ffill preenche os dados faltantes usando o ultimo valor valido encontrado 
df = df.fillna(method='ffill')
df.head()
```

```python
#podemos usar o replace para substituir valores em conjunto com regex 
df = pd.read_csv("/content/log.csv")
df.head(20)
```

```python
# Para substituir usando um regex, fazemos o primeiro parâmetro para substituir o padrão regex que queremos corresponder, o
# segundo parâmetro o valor que queremos emitir na correspondência e, em seguida, passamos um terceiro parâmetro "regex=True".
```

```python
#imagine que queremos detectar todas as páginas html em
# a coluna "video", digamos que isso significa apenas que eles terminam com ".html", e queremos substituir isso com o
# palavra-chave "webpage". Como poderíamos realizar isso?
df.replace(to_replace=".*.html$", value="webpage", regex=True)
```
